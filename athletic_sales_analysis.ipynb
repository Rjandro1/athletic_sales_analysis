```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv("Resources/athletic_sales_2020.csv")
sales_2021_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
Combined_data_df = pd.concat([sales_2020_df, sales_2021_df], axis=0)
Combined_data_df
```

```python
# Check if any values are null.
Combined_data_df.isnull().values.any()
```

```python
# Check the data type of each column
Combined_data_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype

Combined_data_df['invoice_date'] = Combined_data_df['invoice_date'].astype("datetime64[ns]")
```

```python
# Confirm that the "invoice_date" data type has been changed.
Combined_data_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
Total_products_df = Combined_data_df.groupby(["city", "state", "region"])["units_sold"].sum()
Total_products_df = Total_products_df.reset_index()
Total_products_df.head()

Total_products_df.sort_values(by='units_sold', ascending = False)
Total_Products_Sold = Total_products_df.sort_values(by='units_sold', ascending = False)

# Show the top 5 results.
Total_Products_Sold.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table = pd.pivot_table(Total_Products_Sold, values="units_sold", index=["region"], aggfunc="sum")
pivot_table



# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table.rename(columns={"units_sold": "total_products_sold"}, inplace=True)
pivot_table

# Show the top 5 results.
pivot_table.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table_4 = pd.pivot_table(Womens_footwear_df, values="total_sales", index=["invoice_date"], aggfunc="sum")
pivot_table_4


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

pivot_table_5 = pivot_table_4.groupby(["invoice_date"]).sum()

pivot_table_5
# Sort the resampled pivot table in descending order on "Total Sales".
```

```python
Womens_footwear_df["invoice_date"] = pd.to_datetime(Womens_footwear_df["invoice_date"])
pivot_table_4 = pd.pivot_table(
    Womens_footwear_df,
    values="total_sales",
    index="invoice_date",
    aggfunc="sum"
)
pivot_table_4
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

weekly_sales = pivot_table_4.resample('W').sum()
weekly_sales
# Sort the resampled pivot table in descending order on "Total Sales".
resampled_by_sales = weekly_sales.sort_values(by='total_sales', ascending = False)
resampled_by_sales
```
